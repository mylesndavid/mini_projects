# Generating All Required Images

```python
windll.shcore.SetProcessDpiAwareness(1) # 4k resolution
root = Tk()
root.withdraw()
root.attributes('-topmost', True)
open_file = filedialog.askopenfilename()
```

```python
with open(open_file, encoding="utf8") as f:
    lines = f.read().splitlines()
```

```python
prof_pic_dict = {'Billy': 'profile_pictures/billy.PNG',
                'Shregory': 'profile_pictures/shregory.PNG',
                'Teacher': 'profile_pictures/teacher.PNG',
                'Professor': 'profile_pictures/professor.PNG'}
```

```python
# get current time
current_time = datetime.datetime.now()
```

```python
os.system(f'mkdir chat')
name_up_next = False
current_name = None
current_lines = []
finished_message = False
ims = []
msg_number = 1

for line in lines:
    # If line is empty (signifying end of message chain)
    if line=='':
        name_up_next=True
        current_lines = []
        continue
    # lines that are commented out
    if line[0]=='#':
        continue
    # If currently on a name line
    if name_up_next:
        current_name = line.split(':')[0]
        name_up_next = False
        continue
    # otherwise append message
    current_lines.append(line)
    generate_chat(current_lines,
                  current_name,
                  f'{current_time.hour % 12}:{current_time.minute}',
                  prof_pic_dict[current_name]).save(f'{path}\\chat\\{msg_number:03d}{current_name[0]}.png')
    # update time and msg_number for saving
    current_time+=datetime.timedelta(0,30)
    msg_number+=1
    
```
